```python
sns.countplot(data['Ship Mode'])
plt.show()
```

```python
sns.countplot(data['Segment'])
plt.show()
```

```python
data.hist(figsize=(10,10),bins=30)
plt.show()
```

```python
plt.figure(figsize=(12,17))
plt.scatter(x=data['Sales'],y=data['Profit'],c='m')
plt.title('Sales Vs Profit')
plt.xlabel('Total Sale')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(10,8))
data["Sub-Category"].value_counts().plot.pie(autopct= '%1.1f%%',)

plt.show()
```

```python
plt.figure(figsize=(11,17))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.xticks(rotation=90)
plt.show()
```
